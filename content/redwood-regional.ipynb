The parameters for the below map are `"observed_on": "2014-08-08"`, retrieved from [my](http://www.inaturalist.org/observations/digitalvapor) and [Fonda](http://www.inaturalist.org/observations/star_bit)'s iNaturalist observations by appending `.json` to the end ([link](http://www.inaturalist.org/observations/digitalvapor.json)). See their [RESTful API](http://www.inaturalist.org/pages/api+reference). From an [observations endpoint](http://www.inaturalist.org/pages/api+reference#get-observations) you can get 100 results, after that you'll need to authenticate. I set our names in the `pelicanconf.py` file with `NATURALISTS = ['digitalvapor','star_bit']`

* https://github.com/danielfrg/pelican-ipynb
* http://www.inaturalist.org/pages/api+reference
* http://www.inaturalist.org/observations/digitalvapor.json

````python
import requests
import folium
date = '2014-08-08'
usernames = ['digitalvapor','star_bit']
#check if showing map by a day's adventure, if no date then it shows user info
try:
    date
except:
    for user in usernames:
        #if no date, just return user info for now
        url = 'http://www.inaturalist.org/people/'+user+'.json'
        params = ['login','observations_count','identifications_count','id','life_list_taxa_count','user_icon_url']
        r = requests.get(url)
        d = r.json()
        for p in params:
            print(d[p])
else:
    y,m,d=date.split('-')
    filter_payload = {'year':y,'month':m,'day':d}
    num_markers = 0
    for user in usernames:
        url = 'http://www.inaturalist.org/observations/'+user+'.json'
        r = requests.get(url)
        data = r.json()
        for x in range(0,len(data)):
            #creature
            try:
                rank = unicode(data[x]['taxon']['rank'])+' '
            except:
                rank = ''
            try:
                name = unicode(data[x]['taxon']['name'])
            except:
                name = 'something'
            creature = '{0}{1}'.format(rank,name)
            print(creature)
            #geo
            lat = unicode(data[x]['latitude'])
            lon = unicode(data[x]['longitude'])
            position = 'coords: ({0},{1})'.format(lon,lat)
            print(position)
            if(num_markers==0):
                map_rr = folium.Map(location=[lat,lon],zoom_start=14)
            map_rr.simple_marker(location=[lat,lon],popup=creature)
            
            num_markers = num_markers+1
    path='redwood_regional.html'
    map_rr.create_map(path)
````
